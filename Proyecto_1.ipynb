# Modelos de Correlación

Basados en este análisis, existen dos **"x"** con mayor potencial.

Se trabajarán los modelos con estas variables independientes, con la definición de una función de **ENTRENAMIENTO**.

Inicialmente, se define la función que grafica el comportamiento del error entre el modelo de predicción y los valores reales:


```python
def grafico_error(errores):
    plt.scatter(np.array(range(len(errores))), errores,c='green',s=10)
    plt.title('Errores')
    plt.xlabel('epoch')
    plt.ylabel('error')
    plt.show()
```

Se define la función de **ENTRENAMIENTO**:
* Parámetros de Entrada: 
    * Vector "X" 
    * Vector "Y"
    * Número de Iteraciones (EPOCH)
    * Frecuencia de despliegue de errores del modelo. 
    * Learning Rate. 
    * También se incluyen como parámetros opcionales valores iniciales para "m" y "b" (con valors por defecto 0.0)
* Parámetros de Salida: 
    * Diccionario con Historial de Parámetros (m,b) 
    * Vector de Errores.
    * Función del Modelo Entrenado

```python
def entrenamiento(x,y,epoch,imprimir_error_cada,lr, m=0.0, b=0.0):
    
    parametros = np.array([[m],[b]]) # vector (m,b) con valores iniciales
    vector_error=[0.0]*epoch # vector de errores
    diccionario_modelo = {} # inicialización de diccionario para almacenar modelo (m,b)
    
    unos = np.ones((np.size(x), 1))
    matriz_x = np.concatenate((x,unos),axis=1)
    
    
    for i in range(epoch):
        y_hat = np.dot(matriz_x,parametros)  # Producto punto para actualizar el valor de predicción y_hat = mX + b
        error = 0.5 * np.mean((y - y_hat)**2) # cálculo del error
        vector_error[i] = error # almacenamiento del error en un vector
        
        vector_gradiente = np.mean(matriz_x*(y_hat - y),axis=0) # vector_gradiente (gradiente_m, gradiente_b)
        
        # Cálculo de parámetros del modelo
        m = parametros[0] - lr * vector_gradiente[0]       # actualizando m
        b = parametros[1] - lr * vector_gradiente[1]       # actualizando b
        
        # Almacenamiento de historia de parámetros (m,b)
        diccionario_modelo[i] = [np.around(m[0],2),np.around(b[0],2)]
        
        # Actualización de parámetros del modelo para la siguiente iteración
        parametros[0] = m
        parametros[1] = b
    
    
        # Visualización de progresión de errores y sus respectivos parámetros m y b
        if (i+1) % imprimir_error_cada == 0 or i == 0:
            print("EPOCH ",i+1,": Error =",np.around(error,2)," [m,b]=",diccionario_modelo[i])
    
    
    # Función para Gráfica de pérdidas
    grafico_error(vector_error)
    
    print("Estimación Final de parámetros:")
    print("m: ",np.around(parametros[0,0],2))
    print("b: ",np.around(parametros[1,0],2))
    print("pérdida mínima :",np.around(error,2))
    
    def funcion(x):
        return parametros[0,0] * x + parametros[1,0]

    return diccionario_modelo,vector_error,funcion
```

Llamada a la función de **ENTRENAMIENTO** con la primera variable "X" **OverallQual**:

```python
diccionario_OverallQual = {}
error_iteracion_OverallQual = []

diccionario_OverallQual,error_iteracion_OverallQual,manual_OverallQual = entrenamiento(OverallQual,SalePrice,10000,1000,0.00001)
```

Llamada a la función de **ENTRENAMIENTO** con la primera variable "X" **1stFlrSF**:

```python
diccionario_FstFlrSF = {}
error_iteracion_FstFlrSF = []

diccionario_FstFlrSF,error_iteracion_FstFlrSF,manual_FstFlrSF = entrenamiento(FstFlrSF,SalePrice,1000,100,0.00000001)
```

## Evolución de Modelos
Se crea una función para observar la evolución del modelo en el tiempo.
La función grafica la evolución del modelo cada "n" iteraciones, y las grafica en conjunto con los datos de entrenamiento del modelo.

* Parámetros de entrada:
    * Diccionario de valores históricos del modelo (valores [m,b] del modelo y = mx + b.
    * Vector de Valores "X"
    * Vector de Valores "Y" de entrenamiento
    * Frecuencia de Iteración a Mostrar (se muestra el modelo cada "n" valores dentro del Diccionario).

```python
def evolucion(diccionario,x,y,n):
    
    modelo_i = np.empty_like(x)
    
    print("Entrenamiento:")
        
    for i in range(len(diccionario)): # Para cada iteración en el Diccionario de Parámetros m,b
        
        if (i+1) % n == 0 or i == 0: # Cada "n" iteraciones...
            
            for j in range(len(x)): # Se calcula el modelo: y_hat's correspondientes para todas las "x"
                modelo_i[j] = diccionario[i][0]*x[j] + diccionario[i][1]   # y_hat = m*x + b                
            
            # Se muestran detalles del modelo y se dibuja la recta con Matplotlib
            print("Iteración",i+1,": y = ",diccionario[i][0],"*x + ",diccionario[i][1])           
            plt.plot(x,modelo_i,label='Iteración No. {i}'.format(i=i+1))
            plt.legend(loc = 'best')
                
    print("")
    plt.scatter(x,y,c='olive',s=10)
    plt.title('Evolución del Entrenamiento')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    return
```

# Entrenamiento con Scikit-Learn

A continuación se muestra el entrenamiento con herramientas de Scikit-Learn para los modelos que se entrenaron de forma manual anteriormente.
Específicamente los modelos:
* SalesPrice = m * OverallQual + b
* SalesPrice = m * 1stFlrSF + b

Inicialmente se define la función que calculará el modelo lineal sobre cualquier "X" utilizando Scikit-Learn:

```python
from sklearn import linear_model

def modelo_scikit(x,y):
    modelo = linear_model.LinearRegression() # Creación del modelo lineal
    modelo.fit(x, y) # Entrenamiento del modelo con la función Fit
    
    print("Parámetros del modelo: m = ",np.around(modelo.coef_[0,0],2),", b = ", np.around(modelo.intercept_[0],2))
    
    def funcion(x):
        return modelo.coef_[0,0] * x + modelo.intercept_[0] # Modelo lineal con Scikit-Learn
    
    return funcion
```

# Predicción Manual, Scikit-Learn y Promedio

A continuación se creará una función que promediará los modelos obtenidos con anterioridad: Modelo Manual y Modelo Scikit-Learn.

```python
def modelo_promedio(funcion_manual,funcion_scikit,x):

    prediccion_manual=np.empty_like(x)
    prediccion_scikit=np.empty_like(x)
    prediccion_promed=np.empty_like(x)

    for i in range(len(x)):
        prediccion_manual[i] = funcion_manual(i)
        prediccion_scikit[i] = funcion_scikit(i)
    
    # Cálculo vectorizado para el modelo promedio
    prediccion_promed = (prediccion_manual + prediccion_scikit)/2.0;
    
    plt.plot(x,prediccion_manual,label='Prediccion Manual')
    plt.plot(x,prediccion_scikit,label='Prediccion Scikit-Learn')
    plt.plot(x,prediccion_promed,label='Prediccion Promedio')
    plt.legend(loc = 'best')
    plt.title('Modelos de Prediccion')
    plt.xlabel('x')
    plt.ylabel('y_hat')
    plt.show()
    
    return prediccion_manual,prediccion_scikit,prediccion_promed
```

### Predicciones sobre OverallQual (TEST)
Con el objetivo de modularizar el código y ordenar los resultados se crearán dos funciones:
* Función para el Cálculo de Errores: Recibe dos modelos y calcula el error para cada uno de los modelos y el error para el modelo promedio.
* Función para graficar barras agrupadas.

Primero se define la función de cálculo **vectorizado** de errores para los modelos Manual, Scikit y Promedio:
* $ error = \frac{1}{2} * promedio ((y-yhat)^2)$

```python
def errores_prueba(fun_manual,fun_scikit,x_test,y_test):
     
    error_modelo_manual=np.around (0.5 * np.average( (y_test-fun_manual(x_test))**2),3)
    error_modelo_scikit=np.around (0.5 * np.average( (y_test-fun_scikit(x_test))**2),3)
    error_modelo_promed=np.around (0.5 * np.average( (y_test-((fun_manual(x_test)+fun_scikit(x_test))/2.0))**2),3)


    return [error_modelo_manual, error_modelo_scikit, error_modelo_promed]
```

Cálculo de errores para la "X" **OverallQual**:

```python
OverallQual_errores_prueba=errores_prueba(
        manual_OverallQual,scikit_OverallQual,OverallQual_Test,SalePrice_Test)
print(OverallQual_errores_prueba)
```

Cálculo de errores para la "X" **1stFlrSF**:

Función para la agrupación y graficación de errores:

```python
def graficar_errores(lista_errores1,lista_errores2):
    
    # Lista de valores de errores a graficar (características por grupo)
    Manual_errores = [lista_errores1[0],lista_errores2[0]]
    Scikit_errores = [lista_errores1[1],lista_errores2[1]]
    Promed_errores = [lista_errores1[2],lista_errores2[2]]

    # Gráfica de Barras
    fig,barras = plt.subplots()
    
    # Características de Barras
    ancho_barra = 0.25 # Ancho de las barras
    etiquetas = ['OverallQual', '1stFlrSF'] # Nombre de los grupos de barras
    posicion = np.arange(len(etiquetas)) # Posición de la etiqueta del grupo

    
    # Cálculo de grupos de barras
    grupo1 = barras.bar(posicion - ancho_barra, Manual_errores, ancho_barra, label='Manual') # Grupo Modelo Manual
    grupo2 = barras.bar(posicion , Scikit_errores, ancho_barra, label='Scikit') # Grupo Modelo Scikit-Learn
    grupo3 = barras.bar(posicion + ancho_barra, Promed_errores, ancho_barra, label='Promedio') # Grupo Modelo Promedio  
    

    # Etiquetas de la gráfica
    barras.legend()
    barras.set_title('Errores por Modelo y "X"')
    barras.set_ylabel('Errores')   
    barras.set_xticks(posicion)
    barras.set_xticklabels(etiquetas)
    

    # Función que añade anotaciones a cada grupo de barras
    def etiqueta_barras(grupo):
        for i in grupo:
            altura = i.get_height() # Se determina la altura de la barra
            barras.annotate('{}'.format(np.format_float_scientific(altura,2)), # Altura de Barras
                        size=8, # Tamaño de la fuente
                        xy=(i.get_x() + i.get_width() / 2, altura), # Coordenadas de la etiqueta
                        xytext=(0, 3), # Altura sobre el borde de la barra
                        textcoords="offset points",
                        ha='center', va='bottom')

    # Etiquetado de características en cada grupo
    etiqueta_barras(grupo1) # Altura de barras Modelo Manual
    etiqueta_barras(grupo2) # Altura de barras Modelo Scikit-Learn
    etiqueta_barras(grupo3) # Altura de barras Modelo Promedio

    # Se grafica
    plt.yscale("log") # Se grafica en escala logarítmica el eje "y con el propósito de ver claramente ambos grupos"
    plt.show()
    
    return
```

```python
graficar_errores(OverallQual_errores_prueba,FstFlrSF_errores_prueba)
```
